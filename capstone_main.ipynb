```python
def normalize_time(es_t):
    '''
    This function changes time from H.M.S format to S format

    '''
            
    h, m, s = es_t.split('.')
    t_in_sec  = int(h) * 3600 + int(m) * 60 + int(s)

    
    return  t_in_sec   
```
